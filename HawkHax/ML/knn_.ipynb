```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
```

```python
df = pd.read_csv('data.csv')
df.columns = df.columns.str.strip()
df.pop("Unnamed: 0")
df.head(5)
```

```python
def find_nearest_neighbor(email):
    correct_index = df.index[df['Email'] == email]
    print(correct_index)

    arr = df.loc[:, df.columns != 'Email'].to_numpy()


    diff = arr[correct_index] - arr
    np.abs(diff)

    distances = np.sum(diff, axis = 1)
    
    distances_corrected = np.delete(distances, correct_index)
    return distances_corrected.argmin()
```

```python
x = find_nearest_neighbor('0@lol.com')
x
```
